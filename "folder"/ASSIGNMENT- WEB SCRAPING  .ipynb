```python
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
import regex as re
```

```python
url='https://www.imdb.com/list/ls056092300/'
headers={'User-Agent':'Mozilla/5.0'}
# send request to the web page
web=requests.get(url,headers=headers)
print(web)
```

```python
soup=BeautifulSoup(web.content,'html.parser')
```

```python
movie_name=soup.find('div',class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title")
print(movie_name.text)
```

```python
rating=soup.find('div',class_="sc-e2dbc1a3-0 ajrIH sc-b189961a-2 fkPBP dli-ratings-container")
print(rating.text.replace("\xa0",""))
```

```python
year=soup.find('div',class_="sc-b189961a-7 feoqjK dli-title-metadata")
Relese_year= year.text
pattern=r'(\d{4}|\d{2}\w{1})'
year_release=re.sub(pattern,r"\1 ",Relese_year)
print(year_release)
```

```python
movie_names = []
for mov in soup.find_all('div',class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title"):
    movie_names.append(mov.text)
print(movie_names)
```

```python
movie_ratings = []
for rat in soup.find_all('div',class_="sc-e2dbc1a3-0 ajrIH sc-b189961a-2 fkPBP dli-ratings-container"):
    movie_ratings.append(rat.text.replace("\xa0",''))
print(movie_ratings)
```

```python
movie_years = []
for yer in soup.find_all('div', class_="sc-b189961a-7 feoqjK dli-title-metadata"):
    movie_years.append(yer.text)
    print(movie_years)
```

```python
years_with_spaces = []
for year in movie_years:
    years_with_spaces.append(re.sub(r'(\d{4}|\d{2}\w{1})', r'\1 ', year))
print(years_with_spaces)
```

```python
df=pd.DataFrame({'Movie Name':movie_names,'Movei Rating':movie_ratings,'Relase year':years_with_spaces})
print(df)
```

```python
url = 'https://www.patreon.com/coreyms'

# Fetch the content of the page
response = requests.get(url)
html_content = response.text

# Parse HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all the posts
posts = soup.find_all('div', class_="sc-1cvoi1y-0 hxhWXn")

# Iterate over each post
for post in posts:
    # Extract heading
    heading = post.find('a',href="/posts/python-tutorial-103296452").text.strip()
    
    # Extract date
    date = post.find('time').text.strip()
    
    # Extract content
    content = post.find('div', class_='postCard__excerpt').text.strip()
    
    # Extract likes (if available)
    likes_elem = post.find('button', class_='postActions__likeButton')
    likes = likes_elem.find('span', class_='postActions__likeCount').text.strip() if likes_elem else 'No likes'
    
    print(f"Heading: {heading}")
    print(f"Date: {date}")
    print(f"Content: {content}")
    print(f"Likes: {likes}")
    print()
```

# 4 Question Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular .

```python
def scrape_ten_product_details():
    url='https://www.bewakoof.com/bestseller?sort=popular%20'
    response_web=requests.get(url)
    response_web
    soup=BeautifulSoup(response_web.content,'html.parser')
    #print(soup)
    
    #scrape Product Details
    
    Product_name=[nam.text for nam in soup.find_all('div', class_="productNaming bkf-ellipsis")[:10]]
    Price=[pri.text for pri in soup.find_all('div',class_="loyaltyPriceBox")[:10]]
    image_url=[img['src'] for img in soup.find_all('img',class_="productImgTag")[:10]]
    
    #dataFrame
    
    df_bewakoof=pd.DataFrame({'Product name':Product_name,'Product Price':Price,'Product Images':image_url})
    return df_bewakoof

scrape_ten_product_details()
    
```

```python
soup=BeautifulSoup(response.content)
```

```python
headings=[]
for head in soup.find_all('a',class_="RiverCard-title"):
    headings.append(head.text)
print(headings)
```

```python
date=[]
for dt in soup.find_all('span',class_="RiverCard-time"):
    date.append(dt.text)
print(date)
```

```python
news_links = []
for link in soup.find_all('a', class_='RiverCard-title'):
        news_links.append(link['href'])
print(news_links)
```

```python
df_news=pd.DataFrame({'Headings Of News':headings,'Dtes of News':date,'News link':news_links})
print(df_news)
```

a) Paper title b) date c) Author

```python
url='https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/'
page=requests.get(url)
print(page)
# download web page content
soup=BeautifulSoup(page.content)
print(soup)
```

```python
paper_title=[]
for tit in soup.find_all('h2',class_="h5 article-title"):
    paper_title.append(tit.text.strip())
print(paper_title)
```

```python
date=[]
for dt in soup.find_all('p',class_="article-date"):
    date.append(dt.text)
print(date)
```

```python
Author=[]
for aut in soup.find_all('p',class_="article-authors"):
    Author.append(aut.text)
print(Author)
```

```python
df_paper=pd.DataFrame({'Paper Title':paper_title,'Date Of paper':date,'Author of Paper':Author})
print(df_paper)
```
